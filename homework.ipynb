```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import pandas as pd

data = pd.read_csv('homework.csv')
xdata = data.x.to_numpy().reshape(-1,100)[::3,::3]
ydata = data.y.to_numpy().reshape(-1,100)[::3,::3]
zdata = data.energy.to_numpy().reshape(-1,100)[::3,::3]

plt.figure(dpi=250)
ax = plt.subplot()
ax.scatter(xdata, ydata, c=zdata, cmap='gnuplot', s=8, zorder=2)
plt.contourf(xdata, ydata, zdata, cmap='gnuplot', zorder=1)
plt.colorbar()

plt.title("Potential Energy Surface")
plt.ylabel("y direction")
plt.xlabel("x direction")
plt.show()
```

```python
fig = plt.figure(dpi=200)
ax = fig.add_subplot(projection='3d')

#ax.plot_wireframe(xdata,ydata,zdata, cmap='gnuplot', rstride=1, cstride=1)
surf = ax.plot_surface(xdata, ydata, zdata, cmap='jet', edgecolor='none')
plt.title("Energy")
plt.ylabel("y direction")
plt.xlabel("x direction")
ax.view_init(elev=50, azim=25) 
plt.show()
```
