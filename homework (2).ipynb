```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import pandas as pd

data = pd.read_csv('homework.csv')
xdata = data.x.to_numpy().reshape(-1,100)[::3,::3]
ydata = data.y.to_numpy().reshape(-1,100)[::3,::3]
zdata = data.energy.to_numpy().reshape(-1,100)[::3,::3]

plt.figure(dpi=250)
ax = plt.subplot()
ax.scatter(xdata, ydata, c=zdata, cmap='gnuplot', s=26, zorder=2)
plt.contourf(xdata, ydata, zdata, cmap='gnuplot', zorder=1)
plt.colorbar()

plt.title("Energy showing")
plt.ylabel("y direction")
plt.xlabel("x direction")
plt.show()
```
